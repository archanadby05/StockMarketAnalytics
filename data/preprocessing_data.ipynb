```python
import pandas as pd
```

```python
df = pd.read_csv('nifty50.csv')
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()  # remove spaces
df.columns = df.columns.str.upper()  # convert all to uppercase
```

```python
# Melt into long format
long_df = df.melt(id_vars=['DATE'], var_name='Symbol', value_name='Close')
```

```python
# Save processed CSV
long_df.to_csv('nifty50_long.csv', index=False)
print(long_df.head())
```
